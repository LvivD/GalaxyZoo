```python
from albumentations import (
    RandomRotate90, Flip, Compose, Rotate, Crop
)
import numpy as np

def aug(prob=1):
    return Compose([
        RandomRotate90(p=1*prob),
        Flip(p=0.75*prob),
        Rotate(p=0.75*prob),
        Crop(x_min=149, x_max=245, y_min=149, y_max=245)
    ], p=1)
```

```python
import numpy as np
import pandas as pd
import os
from skimage import io

data_csv = "../data/training_solutions_rev1/training_solutions_rev1.csv"
root_dir = "../data/images_training_rev1"

csv_data = pd.read_csv(data_csv)

res = np.zeros((424,424,3), dtype=np.uint64)

n = len(csv_data)

for i in range(10):
    img_path = os.path.join(root_dir, str(csv_data["GalaxyID"][i]) + ".jpg")
    image = io.imread(img_path)
    
    augmentation = aug(prob=1)    
    augmented = augmentation(**{"image":image})
    image_augmanted = augmented["image"]
    
        
res_all = res
    
res //= n
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
io.imshow(res.astype('uint8'))
plt.show()
```
